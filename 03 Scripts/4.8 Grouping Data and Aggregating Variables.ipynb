## 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

To create a loyalty flag for existing customers using transform and loc function:

Step 1
Create a new column, “max_order,” into which you’ll place the results of your aggregation.
df_ords_prods_merge['max_order'] 

Step 2
Split the data into groups based on the “user_id” column
 df_ords_prods_merge.groupby(['user_id'])

Step 3
Apply the transform() function on the “order_number” column to generate the maximum orders for each user.
['order_number'].transform(np.max)


```python
#create a loyalty flag with transform and loc function
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#checking the output
df_ords_prods_merge[['user_id', 'order_number','max_order']].head(15)
```

## flag criteria

```python
#If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer"
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
#If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
#If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#checking the frequency of loyalty flag
df_ords_prods_merge['loyalty_flag'].value_counts(dropna=False)
```

```python
#check output
#to check mulitple columns at ones
df_ords_prods_merge[['user_id', 'loyalty_flag', 'max_order']].head(30)
```

df_ords_prods_merge['Avg_spending']

 df_ords_prods_merge.groupby(['user_id'])

## 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:

    If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
    If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
    If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.

step 1 Create a new column, “User_ord_median” into which you’ll place the results of your aggregation. df_ords_prods_merge['user_ord_median']

Step 2 Split the data into groups based on the “user_id” column df_ords_prods_merge.groupby(['user_id'])

Step 3 Apply the transform() function on the “days_since_prior_order” column to generate the median of user ordering behaviour  for each user. ['days_since_prior_order'].transform(np.median)

```python
#creating user_ordering_behaviour column
df_ords_prods_merge['user_ord_median']= df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#check output
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'user_ord_median']].head(10)
```

## flag criteria

```python
# If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
df_ords_prods_merge.loc[df_ords_prods_merge['user_ord_median']>20, 'frequency_flag']= 'Non-frequent customer'
```

```python
#If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
df_ords_prods_merge.loc[(df_ords_prods_merge['user_ord_median']>10) & (df_ords_prods_merge['user_ord_median']<=20),'frequency_flag']= 'Regular customer'
```

```python
#If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.
df_ords_prods_merge.loc[df_ords_prods_merge['user_ord_median']<=10, 'frequency_flag']= 'Frequent customer'
```

```python
#check the frquency of frequency_flag
df_ords_prods_merge['frequency_flag'].value_counts (dropna=False)
```

```python
#check the output
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'user_ord_median', 'frequency_flag']].head (100)
```
