The customers data is currently a pandas DataFrame, I will convert it to Polars so that it is in the same format as orders_products_merge.

```python
#converting pandas df to polars df
df_custs = pl.from_pandas(df_custs)
```

```python
#Merge (join) df_ords_prods_merge with customers data frame on user_id

df_ords_prods_merge_custs =df_ords_prods_merge.join(df_custs, on="user_id", how="inner")
```
