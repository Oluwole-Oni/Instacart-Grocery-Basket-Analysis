The suitable way to merge the df_ords_prods_combined dataframe with the products dataframe will be to use the 'product_id' column.

```python
#test the merge wihtout overwriting
pd.merge(orders_products_combined, df_prods, on = ['product_id'], indicator='Match')
```

I could not use indicator=True in the merge test above because the orders_products_combined DataFrame already contains a _merge column.
Instead, I used indicator='match'. To enable the use of indicator=True in the new combined DataFrame, I will need to drop the existing '_merge' column from "orders_products_combined", since the merge operation will generate a new '_merge column.

```python
#drop the merge column in the orders_products_combined DataFrame
orders_products_combined = orders_products_combined.drop(['_merge'], axis=1)
```

```python
#check the output 
orders_products_combined.head()
```

```python
#merge updated DataFrame
df_ords_prods_merge = df_prods.merge(orders_products_combined, on = 'product_id', indicator = True)
```
